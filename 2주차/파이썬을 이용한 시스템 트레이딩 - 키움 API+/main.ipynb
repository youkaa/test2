```python
# 4. 조회요청 (연속 조회)
# * TR별로 한 번에 전달할 수 있는 데이터 갯수가 정해져있음. 그 수보다 데이터가 많을 때 연속조회
while kiwoom.prev_next == '2':
    kiwoom.SetInputValue("종목코드", "069500")
    kiwoom.SetInputValue("기준일자", "20160624")
    kiwoom.SetInputValue("수정주가구분", 1)
    kiwoom.CommRqData("주식일봉차트조회요청", "opt10081", 2, "0101")
```

## 5. 매수하기, 매수하기 ( 9:00 ~ 15:30 장 중에만 동작)
- SendOrder() 함수 호출
- OnReceiveTrData(), OnReceiveMsg(), OnReceiveChejan() 이벤트 차례대로 발생
- OnReceiveTrData() : 주문 번호를 얻을 수 있다. 실패 시 공백("")으로 전달
- OnReceiveMsg() : 주문 성공, 실패 메시지를 코드와 함께 전달
- OnReceiveChejan() : 주문 접수, 체결, 잔고 발생 시 호출, 대부분의 주문 관련 정보를 얻을 수 있다
- *GetChejanData() : OnReceiveChejan() 이벤트 함수가 호출될 때 체결 정보나 잔고정보를 얻어오는 함수

#### 매수하기

```python
# 2. 서버로부터 사용자 조건 검색식 가져오기, 파일로 로컬에 임시 저장한다.
kiwoom.GetConditionLoad()
```

```python
# 3. 임시로 저장한 사용자 조건식 파일에서 조건명리스트를 전부 불러온다.
f = open("condition.txt", "r")
data = f.read()
splited = data.split(";")[:-1]
condition_list = {}
for i in range(len(splited)):
    splited2 = splited[i].split("^")
    condition_list.update({splited2[0]:splited2[1]})
print(condition_list)
```

```python
for key, value in condition_list.items():
    print("조건식명 : " + value)
    kiwoom.SendCondition("0156", value, key, 0)
    print("")
```
